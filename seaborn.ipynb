# Distribution Plots

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df['age']=df['age'].fillna(0)
```

## distplot

```python
sns_plot=sns.distplot(df['age'])
sns_plot.figure.savefig("output.png")
```
